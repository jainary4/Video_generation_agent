```python
# model video generation code without manim rag
%%manim -qm -v WARNING DerivativeExplanation

class DerivativeExplanation(Scene):
    def construct(self):
        # Create axes
        axes = Axes(
            x_range=[-3, 3, 1],
            y_range=[-1, 10, 1],
            axis_config={"color": BLUE},
        )
        axes_labels = axes.get_axis_labels(x_label="x", y_label="f(x)")

        # Define the function f(x) = x^2
        func = axes.plot(lambda x: x**2, color=YELLOW)
        func_label = MathTex("f(x) = x^2").next_to(func, UR)

        # Define the derivative f'(x) = 2x
        derivative = axes.plot(lambda x: 2*x, color=GREEN)
        derivative_label = MathTex("f'(x) = 2x").next_to(derivative, UR)

        # Create a dot that will move along the function
        dot = Dot().move_to(axes.c2p(0, 0))
        dot_label = MathTex("(x, f(x))").next_to(dot, UR)

        # Create a tangent line
        tangent_line = always_redraw(lambda: axes.get_secant_slope_group(
            x=dot.get_center()[0],
            graph=func,
            dx=0.01,
            secant_line_color=RED,
            secant_line_length=3
        ))

        # Animation sequence
        self.play(Create(axes), Write(axes_labels))
        self.play(Create(func), Write(func_label))
        self.wait(1)
        self.play(Create(derivative), Write(derivative_label))
        self.wait(1)
        self.play(FadeIn(dot), Write(dot_label))
        self.play(Create(tangent_line))
        self.wait(1)
        self.play(dot.animate.move_to(axes.c2p(2, 4)), run_time=3)
        self.wait(1)
        self.play(dot.animate.move_to(axes.c2p(-1, 1)), run_time=3)
        self.wait(2)

# To run this script, save it as a .py file and run it using Manim.
```

```python
 # MODEL VIDEO GENERATION CODE WITH MANIM DOCS RAG
 %%manim -qm -v WARNING DerivativeExplanation
 class DerivativesExplanation(Scene):
    def construct(self):
        # Title
        title = Text("Understanding Derivatives", font_size=48)
        self.play(Write(title))
        self.wait(2)
        self.play(FadeOut(title))

        # Introduction to the concept
        intro_text = Text(
            "A derivative represents the rate of change of a function.",
            font_size=32
        )
        self.play(Write(intro_text))
        self.wait(3)
        self.play(FadeOut(intro_text))

        # Example function
        axes = Axes(
            x_range=[-3, 3, 1],
            y_range=[-1, 10, 1],
            axis_config={"color": BLUE},
        )
        graph = axes.plot(lambda x: x**2, color=YELLOW)
        graph_label = axes.get_graph_label(graph, label='x^2')

        self.play(Create(axes), Create(graph), Write(graph_label))
        self.wait(2)

        # Tangent line to illustrate derivative
        tangent_line = axes.get_secant_slope_group(
            x=1,
            graph=graph,
            dx=0.01,
            secant_line_color=GREEN,
            secant_line_length=4
        )
        self.play(Create(tangent_line))
        self.wait(2)

        # Explanation of the tangent line
        derivative_text = Text(
            "The slope of the tangent line at a point gives the derivative at that point.",
            font_size=32
        )
        derivative_text.next_to(tangent_line, DOWN)
        self.play(Write(derivative_text))
        self.wait(3)

        # Conclusion
        conclusion_text = Text(
            "Derivatives are fundamental in calculus for understanding change.",
            font_size=32
        )
        self.play(
            FadeOut(graph),
            FadeOut(axes),
            FadeOut(tangent_line),
            FadeOut(derivative_text)
        )
        self.play(Write(conclusion_text))
        self.wait(3)
        self.play(FadeOut(conclusion_text))
```

```python
%%manim -qm -v WARNING CurvedSurfacesScene
from manim import *
import numpy as np

class CurvedSurfacesScene(ThreeDScene):
    def construct(self):
        # Set up camera orientation for 3D view
        self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)

        # Title at the top
        title = Text("Curved Surfaces", font_size=48)
        title.to_edge(UP)
        self.play(Write(title))
        self.wait(1)

        # ------------------------------
        # 1. Sphere: x+y+z=R with R=2.
        sphere = Sphere(radius=2, resolution=24, color=BLUE)
        sphere_label = Text("Sphere: x+y+z=4", font_size=36)
        sphere_label.to_corner(UL)
        self.play(Create(sphere), Write(sphere_label))
        self.wait(2)
        self.play(FadeOut(sphere), FadeOut(sphere_label))
        self.wait(0.5)

        # ------------------------------
        # 2. Ellipsoid: (x/3)+(y/2)+(z/1.5)=1.
        ellipsoid = Sphere(radius=1, resolution=24, color=GREEN)
        # Stretch the sphere along each axis
        ellipsoid.stretch_to_fit_width(6)    # a=3  width=6
        ellipsoid.stretch_to_fit_height(4)   # b=2  height=4
        ellipsoid.stretch_to_fit_depth(3)    # c=1.5  depth=3
        ellipsoid_label = Text("Ellipsoid: (x/3)+(y/2)+(z/1.5)=1", font_size=28)
        ellipsoid_label.to_corner(UR)
        self.play(Create(ellipsoid), Write(ellipsoid_label))
        self.wait(2)
        self.play(FadeOut(ellipsoid), FadeOut(ellipsoid_label))
        self.wait(0.5)

        # ------------------------------
        # 3. Paraboloid: z = x+y (surface of revolution)
        # Use ParametricSurface if available; otherwise fallback to Surface.
        try:
            paraboloid = ParametricSurface(
                lambda u, v: np.array([u * np.cos(v), u * np.sin(v), u**2]),
                u_range=[0, 2],
                v_range=[0, TAU],
                resolution=(20, 32),
                color=PURPLE,
                opacity=0.8
            )
        except NameError:
            paraboloid = Surface(
                lambda u, v: np.array([u * np.cos(v), u * np.sin(v), u**2]),
                u_range=[0, 2],
                v_range=[0, TAU],
                resolution=(20, 32),
                color=PURPLE,
                opacity=0.8
            )
        paraboloid_label = Text("Paraboloid: z=x+y", font_size=28)
        paraboloid_label.to_corner(DL)
        self.play(Create(paraboloid), Write(paraboloid_label))
        self.wait(2)
        self.play(FadeOut(paraboloid), FadeOut(paraboloid_label))
        self.wait(1)

        # Fade out the title to conclude the scene
        self.play(FadeOut(title))
        self.wait(1)
```

```python
%%manim -qm -v WARNING LinearIndependence
class LinearIndependence(Scene):
    def construct(self):
        # Title
        title = Text("Linear Independence").scale(1.5)
        self.play(Write(title))
        self.wait(2)
        self.play(FadeOut(title))

        # Define vectors
        vector1 = Arrow(start=ORIGIN, end=[2, 1, 0], buff=0, color=BLUE)
        vector2 = Arrow(start=ORIGIN, end=[1, 2, 0], buff=0, color=RED)

        # Labels for vectors
        label1 = MathTex("\\vec{v}_1").next_to(vector1.get_end(), RIGHT)
        label2 = MathTex("\\vec{v}_2").next_to(vector2.get_end(), LEFT)

        # Show vectors
        self.play(GrowArrow(vector1), GrowArrow(vector2))
        self.play(Write(label1), Write(label2))
        self.wait(1)

        # Linear combination
        linear_comb = MathTex("a\\vec{v}_1 + b\\vec{v}_2 = 0").scale(1.2)
        linear_comb.next_to(vector2, DOWN * 2)
        self.play(Write(linear_comb))
        self.wait(2)

        # Explanation
        explanation = Text("Only possible if a = b = 0").scale(0.8)
        explanation.next_to(linear_comb, DOWN)
        self.play(Write(explanation))
        self.wait(3)

        # Fade out
        self.play(
            FadeOut(vector1),
            FadeOut(vector2),
            FadeOut(label1),
            FadeOut(label2),
            FadeOut(linear_comb),
            FadeOut(explanation)
        )
```
